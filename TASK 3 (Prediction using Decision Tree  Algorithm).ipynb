## Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeClassifier
```

## Loading my dataset

```python
iris= pd.read_csv("Iris.csv")
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

## Defining my variables 

```python
iris.columns
```

```python
X = iris[['SepalLengthCm', 'SepalWidthCm',
         'PetalLengthCm', 'PetalWidthCm']]
y = iris['Species'] 
```

## Splitting my data for training and testing

```python
dt=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

## Training my model

```python
dt.fit(X_train,y_train)
```

## Making predictions ffrom my test values

```python
y_pred = dt.predict(X_test)
```

## Comparing my model's prediction to the actual values 

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

## Visualizing my Decision Tree

```python
from sklearn import tree
```

```python
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(dt,filled=True,feature_names=[ 'SepalLengthCm', 'SepalWidthCm','PetalLengthCm', 'PetalWidthCm'],
    class_names= sorted(y.unique()),label='all',rounded=True)
plt.title("decision tree")
```
